```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

db=pd.read_csv('StudentsPerformance.csv')
db
```

```python
#1
sns.countplot(x='gender',data=db)
print("No of male students:",db['gender'].value_counts()['male'])
print("No of female students:",db['gender'].value_counts()['female'])
```

```python
#2
sns.countplot(x='parental level of education',data=db)
plt.title('parental level of education')
plt.xticks(rotation=90)
db['parental level of education'].value_counts()
```

```python
#3a.
avrg=db.groupby('gender')['math score','reading score','writing score'].mean()
print('scores the most on average is:',avrg)
plt.title('Average score wrto gender')
plt.xticks(rotation=90)
ax=sns.lineplot(data=avrg,markers=True)
ax.set(ylabel='Average Score')
```

```python
#3b.
avrg_2=db.groupby('test preparation course')['math score','reading score','writing score'].mean()
print('scores the most on average is:',avrg_2)
plt.title('Average score wrto test preparation corse')
plt.xticks(rotation=90)
ax=sns.lineplot(data=avrg_2,markers=True)
ax.set(ylabel='Average Score')
```

```python
#4a.
scrvar_1=db.groupby('gender')['math score','reading score','writing score'].std()
print('score variation wrto gender is:',scrvar_1)
plt.title('scoring variation wrto gender')
plt.xticks(rotation=90)
ax=sns.lineplot(data=scrvar_1,markers=True)
ax.set(ylabel='Average Score')
```

```python
#4b.
scrvar_2=db.groupby('test preparation course')['math score','reading score','writing score'].std()
print('score variation wrto test preparation course is:',scrvar_2)
plt.title('scoring variation wrto test preparation course ')
plt.xticks(rotation=90)
ax=sns.lineplot(data=scrvar_2,markers=True)
ax.set(ylabel='Average Score')
```
