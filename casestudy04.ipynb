# 3. Check whether all the zones generate the same amount of sales.
H0 -μA = μB = μC = μD(where μ = mean)

Ha - At least one difference among the mean sales.

```python
from scipy.stats import f_oneway
f_oneway(db['Zone - A'], db['Zone - B'], db['Zone - C'],db['Zone - D'])
```

# The F test statistic is 5.672056106843581 and the p-value is 0.0011827601694503335 then the p-value is less than .05, we reject the null hypothesis.Four zones has different sales.
